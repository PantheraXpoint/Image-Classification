```python
def get_cropped_images(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,1.3,5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h,x:x+w]
        roi_color = img[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >=2:
            return roi_color
    
```

```python
path_to_data = './images_dataset/'
path_to_cropped_data = "./images_dataset/cropped/"
```

```python
import os
img_dirs = []
for entry in os.scandir(path_to_data):
    print (entry)
    if entry.is_dir():
        img_dirs.append(entry.path)
```

```python
import shutil
if os.path.exists(path_to_cropped_data):
    shutil.rmtree(path_to_cropped_data)
os.mkdir(path_to_cropped_data)
```

```python
cropped_image_dirs = []
celeb_file_names_dict = {}
for img_dir in img_dirs:
    count = 0
    celeb = img_dir.split('/')[-1]
    celeb_file_names_dict[celeb] = []
    for entry in os.scandir(img_dir):
        ret_img = get_cropped_images(entry.path)
        if ret_img is not None:
            cropped_folder = path_to_cropped_data + celeb
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                
            cropped_file_name = celeb + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name
            cv2.imwrite(cropped_file_path,ret_img)
            celeb_file_names_dict[celeb].append(cropped_file_path)
            count += 1

```

```python
def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
im_har = w2d(img,'db1',5)
plt.imshow(im_har, cmap='gray')
```

```python
class_dict = {}
count = 0
for celeb in celeb_file_names_dict.keys():
    if celeb != 'cropped':
        class_dict[celeb] = count
        count += 1
class_dict
```

```python
X = []
y = []

for celeb, file in celeb_file_names_dict.items():
    if celeb == 'cropped':
        continue
    for img in file:
        image = cv2.imread(img)
        if image is None:
            continue
        scaled_raw = cv2.resize(image,(32,32))
        image_har = w2d(image,'db1',5)
        scaled_har = cv2.resize(image_har,(32,32))
        combined_img = np.vstack((scaled_raw.reshape(32*32*3,1),scaled_har.reshape(32*32,1)))
        X.append(combined_img)
        y.append(class_dict[celeb])
```
